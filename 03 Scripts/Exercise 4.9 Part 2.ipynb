```python
# importing data

ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_customers_merged.pkl'))
```

```python
# creating histogram

ords_prods_merge['order_hour_of_day'].plot.hist(bins=24, color='aqua')
```

```python
# creating bar chart of the loyalty flag

ords_prods_merge['loyalty_flag'].value_counts().plot.bar(color='blue')
```

```python
# creating random seed for accurate sample of data

np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
big = ords_prods_merge[dev]
small = ords_prods_merge[~dev]
```

```python
# checking to make sure the sample data is equal to the whole data

len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
# creating a dataframe to use to find if there is difference in expenditure depending on hour of day

df_2 = small[['order_hour_of_day', 'prices']]
```

```python
# creating line chart for the relationship between 'order_hour_of_day' and 'prices'

line = sns.lineplot(data = df_2, x ='order_hour_of_day', y ='prices')
```

```python
# creating a sample dataframe for 'age' and 'n_dependants'

age_dependants = small[['age', 'n_dependants']]
```

```python
# creating line chart for the relationship between 'age' and 'n_dependants'

line_2 = sns.lineplot(data = age_dependants, x = 'age', y = 'n_dependants')
```

# The chart indicates younger individuals tend to have fewer dependents, while middle-aged customers often have more.

```python
# creating scatterplot for the relationship between 'age' and 'income'

scatter = sns.scatterplot(x = 'age', y = 'income', data = ords_prods_merge)
```

# For a majority of customers there is no correlation between income and age but the chart also tells you that all of the very high end customers (400000 and above) are all over the age of 40
