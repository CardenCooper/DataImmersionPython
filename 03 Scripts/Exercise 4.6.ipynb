```python
# dropping the merge column so that we won't have 2 merge columns after the next merge

orders_products_no_merge = orders_products_combined.drop('_merge', axis = 1 )
```

# Both datasets have a products_id column so we will use that to join


```python
# data sets merge

ords_prods_merge = orders_products_no_merge.merge(df_prods, on = ['product_id'], indicator = True)
```
